```python
import pandas as pd

from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK
# CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts import options as opts
from pyecharts.charts import Bar

from stockdatamanage.db import engine
```

### 读取dv_ttm为空的数据, 时间段20210619-20221108

```python
sql = '''select trade_date, count(1) cnt from daily_basic 
        where trade_date>='20210619' and trade_date<='20221108' and dv_ttm=0
        group by trade_date
        '''
with engine.connect as conn:
        df = pd.read_sql(text(sql), conn)
        print(df.head())
```

```python
dates = df.trade_date.to_list()
dates = [d.strftime('%Y%m%d') for d in dates]
print(dates)
```

```python
bar = (
    Bar()
    .add_xaxis(dates)
    .add_yaxis("数量", df.cnt.to_list())
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Bar-DataZoom（slider-水平）"),
        datazoom_opts=opts.DataZoomOpts(),
    )
)
```

```python
bar.load_javascript()
```

```python
# load_javascript 与 render_notebook方法需在不同的代码格中分别调用
bar.render_notebook()
```
